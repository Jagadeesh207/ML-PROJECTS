```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
scaler=StandardScaler()
```

```python
x=df.drop(columns="target")
```

```python
y=df["target"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
sx_train=scaler.fit_transform(x_train)
```

```python
sx_test=scaler.transform(x_test)
```

```python
sx_test
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
models={
    "Logistic Regression":LogisticRegression(),
    "KNN":KNeighborsClassifier(),
    "Random Forest Classifier":RandomForestClassifier()
}
```

```python
def fit_and_score(models,sx_train,sx_test,y_train,y_test):
    np.random.seed(42)
    model_scores={}
    for name,model in models.items():
        model.fit(sx_train,y_train)
        print(name,":",model.score(sx_test,y_test))
        model_scores[name]=model.score(sx_test,y_test)
    return model_scores
```

```python
model_scores=fit_and_score(models,sx_train,sx_test,y_train,y_test)
```

```python
model_scores
```

```python
model_compare=pd.Series(model_scores)
```

```python
model_compare.plot(kind='bar')
plt.xticks(rotation=0)
plt.show()
```
